```python
'''Its a good exemaple to get 
the Net payable amount by setting 
up discount at specific price range by using if-else statement'''
total_price = (int)(input("Price :"))
if total_price<100:
    discount = total_price*0.04
    print("Discount :" + (str)(discount))
else:
    discount = total_price*0.10
    print("Discount :"+(str)(discount))
print("Net payable amount:",total_price-discount)
    
    
```

#While Loop

```python
'''A child is learning to count number of 10 rupess in his hand
and here we have given 5 notes of 10 rupees.
Output will display the counting of notes and total amount'''
notes=5;
i = 1
while i<=notes:
    print(i)
    i+=1
print("Total Amount",notes*10,"rupees")
```

```python
'''Continuing with same example
here we using else statement'''
i=1
notes=5
while i<=notes:
    print(i)
    i+=1
else:
    print("No 10 rupee notes left")
print("Total money:",10*notes,"rupees")
```

#For Loop

```python
'''A simple example has taken where
we have several fruits in a list and we display
this fruit as per the arrangement in list '''
fruits_list = ["Mango","Cherry","Apple","Papaya","Banana"]##List containing fruits names
for i in fruits_list:
    print(i)
```

#Nested Loops

```python
'''Here we use nested for loop to create a triangle'''
n=7 #n is length and width of triangle
for i in range(0,n): #loop to handel the rows 
    for j in range(0,i+1): #loop to handel the columns
        print("* ",end="")#printing the stars
    print("\r") #after each row line will end
```

```python
'''This example shows the use of nested while loop to
create a pyramid like structure'''
total_row = int(input("Enter the total number of row :"))
row = 0
while row<total_row:
    row=row+1
    space = total_row - row
    sc = 0
    while(sc<space):
        print(" ",end='')
        sc+=1
    star = 2*row-1
    while(star>0):
        print("*",end='')
        star-=1 
    print()
```

#Break statement

```python
'''It is a simple example of break,
where it immediately terminate an operation of
statament body when x is equal to 4 '''
x = 1
while x<7:
    print(x)
    if x == 4:
        break
    x+=1
```

```python
'''Here it terminates the
operational body when fruit name is papaya'''
fruit_list = ["apple","mango","banana","papaya"]
for i in fruit_list:
    if i == "papaya":
        break
    print(i)
```

#Continue statement

```python
'''Is is a simple example of continue,
where it simply terminate an current iteration
when x is equal to 4 and
continues with next iteration'''
x=0
while x<7:
    x+=1
    if x==4 :
        continue
    print(x)
```

```python
fruit = ["banana","mango","papaya"]
for f in  fruit:
    if f == "mango":
        continue
    print(f)
```
